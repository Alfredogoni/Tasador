```python
#Definimos una funcion para crear un GeoDataFrame a partir de un DF con polygonos
def from_wkt(df, wkt_column):
    import shapely.wkt
    df["coordinates"]= df[wkt_column].apply(shapely.wkt.loads)
    gdf = geopandas.GeoDataFrame(df, geometry='coordinates')
    return gdf
```

```python
#Definimos una funcion para crear un GeoDataFrame a partir de un DF con longitud y Latitud
def from_x_y(df, x, y):
    from shapely.geometry import Point
    gdf = geopandas.GeoDataFrame(df,
                                crs={'init': 'epsg:4326'},
                                geometry=[Point(xy) for xy in zip(df[x], df[y])])
    return gdf
```

```python
#Definimos una funcion para ajustar el CRS al utilizado por Data Buenos Aires
def to_gkba(geodf):
    geodf.crs = {'init' :'epsg:4326'}
    new_geodf = geodf.to_crs(crs = "+proj=tmerc +lat_0=-34.629269 +lon_0=-58.4633 +k=0.9999980000000001 +x_0=100000 +y_0=100000 +ellps=intl +units=m +no_defs")
    return new_geodf
```

```python
#Definimos los GeoDataFrames
df_bancos = from_x_y(df_bancos,'X','Y')
df_clubes = from_x_y(df_clubes,'long','lat')
df_comisarias = from_x_y(df_comisarias,'long','lat')
df_bomberos = from_x_y(df_bomberos,'long','lat')
df_espaciov = from_wkt(df_espaciov,'WKT')
df_colegios = from_x_y(df_colegios,'long','lat')
df_restaurants = from_x_y(df_restaurants,'long','lat')
df_subte = from_x_y(df_subte,'long','lat')
df_gym = from_x_y(df_gym,'lng','lat')
df_hospitales = from_x_y(df_hospitales,'long','lat')
df_colectivos = from_x_y(df_colectivos,'X','Y')
df_universidades = from_wkt(df_universidades,'WKT_gkba')
df_crimen = from_x_y(df_crimen,'LONGITUD','LATITUD')
```

```python
#Definimos una funcion para generar un radio alrededor de cada departamente
def buff_x_y(df, x, y):
    from shapely.geometry import Point
    gdf = geopandas.GeoDataFrame(df,
                                crs={'init': 'epsg:4326'},
                                geometry=[Point(xy).buffer(0.005) for xy in zip(df[x], df[y])])
    return gdf
```

```python
#Definimos una funcion para calcular la distancia de los centroides de cada poligono y cada departamento
def nearest_point_pol(geodf, points):
    from shapely.geometry import Point
    dist_depto=[]
    for depto in geodf.geometry:
        distances = [ (i, depto.distance(pt))  for i, pt in enumerate(points.geometry.centroid)]
        dist = [dist for i, dist in distances]
        dist.sort()
        dist_depto.append(dist[0])
    return dist_depto
```

```python
#Definimos una funcion para calcular la distancia de puntos de interes y cada departamento
def nearest_point(deptos, points):
    from shapely.geometry import Point
    dist_depto=[]
    for a,depto in enumerate(deptos.geometry):
        distances = [ (i, pt.distance(depto))  for i, pt in enumerate(points.geometry)]
        dist = [dist for i, dist in distances]
        dist.sort()       
        dist_depto.append(dist[0])
    return dist_depto
```
